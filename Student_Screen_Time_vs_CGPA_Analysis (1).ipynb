# Student Screen Time vs CGPA Analysis
### Complete Project Notebook

This project analyzes the relationship between daily screen time and students' academic performance (CGPA).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Dataset

```
# Update the file name if needed
df = pd.read_csv('student_screen_time.csv')
df.head()
```

## Dataset Information

```
df.info()
```

## Statistical Summary

```
df.describe()
```

## Checking Missing Values

```
df.isnull().sum()
```

## Screen Time vs CGPA Scatter Plot

```
plt.figure()
plt.scatter(df['daily_screen_time_hours'], df['current_sem_CGPA'])
plt.xlabel('Daily Screen Time (Hours)')
plt.ylabel('Current Semester CGPA')
plt.title('Student Screen Time vs CGPA Analysis')
plt.show()
```

## Scatter Plot with Regression Line

```
x = df['daily_screen_time_hours']
y = df['current_sem_CGPA']

plt.figure()
plt.scatter(x, y)

# Regression Line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)

plt.xlabel('Daily Screen Time (Hours)')
plt.ylabel('Current Semester CGPA')
plt.title('Screen Time vs CGPA with Trend Line')
plt.show()
```

## Correlation Analysis

```
df.corr()
```

## Conclusion
- Higher screen time shows a negative impact on CGPA.
- Students with lower screen time generally achieve higher academic performance.
- Balanced digital usage is important for academic success.
